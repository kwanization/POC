#연구자료 불러오기

```python
import numpy as np
import pandas as pd
```

```python
dfs = []

  #파일 번호대로 불러오기
for i in range(1,18):  # 1부터 17까지 파일 불러오기
    file_name = f'연구자료 ({i}).xlsx'

  # 엑셀 파일을 읽어서 리스트에 추가
    df = pd.read_excel(file_name)
    dfs.append(df)

# 세로로 연결
df_combined = pd.concat(dfs, axis=0)

# 데이터 확인
df_combined.head()
```

#중복데이터 제거

```python
#'No' 기준으로 중복 행 삭제
df_combined_cleaned = df_combined.drop_duplicates(subset=['No'], keep='first')

#데이터 확인
df_combined_cleaned.head()

#연구자료 엑셀 파일로 저장
df_combined_cleaned.to_excel('cleaned_data.xlsx', index=False)

#엑셀 파일 다운로드
files.download('cleaned_data.xlsx')

```

```python
#cleaned_data 길이보기
print(len(df_combined))
print(len(df_combined_cleaned))
```

```python
# No 칼럼의 값을 문자열로 변환하고, 앞에 '0000' 붙이기
df_combined_cleaned.loc[:, 'No'] = df_combined_cleaned['No'].astype(str).str.zfill(10)

#데이터 내림차순
df_combined_cleaned = df_combined_cleaned.sort_values(by='No', ascending=False)

# 데이터 확인
df_combined_cleaned.head()
df_combined_cleaned.tail()
```

```python
#No칼럼 값 리스트로 변환
no_list = df_combined_cleaned['No'].tolist()

#txt파일로 저장(리스트 형태로 변환하여 기록)
with open('no_list.txt', 'w') as f:
    f.write(str(no_list))

#txt파일 다운로드
files.download('no_list.txt')
```
